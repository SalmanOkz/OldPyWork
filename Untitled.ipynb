```python
msg = input("Enter your message: ")
shift = int(input("By how many characters you want to shift your letters: "))

encrypted_msg = ""  # Initialize an empty string to store the encrypted message

for letter in msg:
    if letter.isalpha():
        letters = "abcdefghijklmnopqrstuvwxyz"
        if letter.isupper():
            letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        position = letters.find(letter)
        shifted_position = (position + shift) % 26
        encrypted_letter = letters[shifted_position]
        encrypted_msg += encrypted_letter
    else:
        encrypted_msg += letter  # Keep non-alphabetical characters unchanged

print("Encrypted message:", encrypted_msg)

decrypted_msg=""

for letter in encrypted_msg:
    if letter.isalpha():
        letters = "abcdefghijklmnopqrstuvwxyz"
        if letter.isupper():
            letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        position = letters.find(letter)
        print(position)
        shifted_position = (position - shift)
        print(shifted_position)
        encrypted_letter = letters[shifted_position]
        decrypted_msg += encrypted_letter
    else:
        decrypted_msg += letter  # Keep non-alphabetical characters unchanged

print("decrypted_msg message:", decrypted_msg)
```
