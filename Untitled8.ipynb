```python
subject_scores = [('Mathematics', 95), ('Physics', 85), ('Literature', 92), ('Computer Science', 88)]

# Sort the list of tuples based on the scores using a lambda function
sorted_scores = sorted(subject_scores, key=lambda x: x[1], reverse=True)

print("Sorted scores:")
for subject, score in sorted_scores:
    print(f"{subject}: {score}")
```

```python
subject_scores = [('Mathematics', 95), ('Physics', 85), ('Literature', 92), ('Computer Science', 88)]

# Increase all the scores by 5 using the map function
increased_scores = list(map(lambda x: (x[0], x[1] + 5), subject_scores))

print("Increased scores:")
for subject, score in increased_scores:
    print(f"{subject}: {score}")
```

```python
subject_scores = [('Mathematics', 95), ('Physics', 85), ('Literature', 92), ('Computer Science', 88)]

# Filter out and display the subjects where the scores are greater than 90
high_scores = list(filter(lambda x: x[1] > 90, subject_scores))

print("Subjects with scores greater than 90:")
for subject, score in high_scores:
    print(f"{subject}: {score}")
```

```python
subject_scores = [('Mathematics', 95), ('Physics', 85), ('Literature', 92), ('Computer Science', 88)]

# List subjects with scores greater than 85 using list comprehension
strong_subjects = [subject for subject, score in subject_scores if score > 85]

print("Subjects with scores greater than 85:")
for subject in strong_subjects:
    print(subject)
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)



stack = Stack()


stack.push("Dark Chocolates")
stack.push("Milk Chocolates")

print("Stack:", stack)


removed_chocolate = stack.pop()
print("Removed:", removed_chocolate)
print("Stack:", stack)
```

```python
from collections import deque

class Queue:
    def __init__(self):
        self.items = deque()

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.popleft()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)



queue = Queue()


queue.enqueue("Assorted Chocolates")
queue.enqueue("Truffles")

print("Queue:", queue)

removed_chocolate = queue.dequeue()
print("Removed:", removed_chocolate)
print("Queue:", queue)
```
