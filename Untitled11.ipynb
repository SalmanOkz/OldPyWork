```python
def player(vip):
    name=input("Enter player Name: ")
    left_write=input("enter side of your batting leftiiii or writeee")
    unique_number=input("enter unique number: ")
    player = {
        "name" : name,
        "batting style": left_write,
        "unique_number": unique_number
    }
    vip[unique_number]=player
    print("player info added: ")

def delete_player(vip):
    unique_number=input("Enter your player unique number: ")
    if(player is not vip):
        print("not found")
        del vip[unique_number]
        print("player gone on duck:")

vip = {}
while True:
    print("2. add player:")
    print("3. remove player: ")
    print("4. Exit")
    choice=input("Enter any of your choice")
    if choice == 2:
        player(vip)
    elif choice == 3:
        delete_player(vip)
    elif choice == 4:
        break;
    else:
        print("Enter correct value")
        

```
