```python
def book(catalog):
    title = input("Add book of your title: ")
    author = input("Add book of your author: ")
    ISBN = input("Add book of your ISBN: ")
    genre = input("Add book of your genre: ")
    book = {
        "title": title,
        "author": author,
        "ISBN": ISBN,
        "genre": genre
    }
    catalog[ISBN]=book
    print(catalog)
    print("book added:")


catalog = {}
book(catalog)
print(catalog)
```

```python
def book(vip):
    author=input("input author name:")
    isbn=input("input isbn")
    book = {
        "author":author,
        "isbn":isbn
    }
    vip[isbn]= book
    


def delete(vip):
    isbn=input("Enter isbn you want to del: ")
    if(isbn is not vip):
        print("not found: ")
    del vip[isbn]
    print("gone mate: ")

vip = {}
while True:
    print("2.add")
    print("3.del")
    choice = input("input your choice: ")
    if(choice == 2):
        book(vip)
    elif(choice == 3):
        book(delete)
    elif(choice == 4):
        break:
    else:
        print("invalid: ")
```
