```python
def add_income(income_list):
    amount = float(input("Enter income amount: "))
    category = input("Enter income category: ")
    income_list.append({'amount': amount, 'category': category})

def add_expense(expense_list):
    amount = float(input("Enter expense amount: "))
    category = input("Enter expense category: ")
    expense_list.append({'amount': amount, 'category': category})

def get_total_income(income_list):
    return sum(item['amount'] for item in income_list)

def get_total_expenses(expense_list):
    return sum(item['amount'] for item in expense_list)

def get_balance(income_list, expense_list):
    return get_total_income(income_list) - get_total_expenses(expense_list)

def get_summary(income_list, expense_list):
    income_by_category = {}
    for item in income_list:
        if item['category'] not in income_by_category:
            income_by_category[item['category']] = 0
        income_by_category[item['category']] += item['amount']
    
    expenses_by_category = {}
    for item in expense_list:
        if item['category'] not in expenses_by_category:
            expenses_by_category[item['category']] = 0
        expenses_by_category[item['category']] += item['amount']
    
    summary = {
        'total_income': get_total_income(income_list),
        'total_expenses': get_total_expenses(expense_list),
        'balance': get_balance(income_list, expense_list),
        'income_by_category': income_by_category,
        'expenses_by_category': expenses_by_category,
    }
    
    return summary

income_list = []
expense_list = []

while True:
    print("\nBudget Tracker")
    print("1. Add Income")
    print("2. Add Expense")
    print("3. View Summary")
    print("4. Exit")
    
    choice = input("Choose an option: ")
    
    if choice == '1':
        add_income(income_list)
    elif choice == '2':
        add_expense(expense_list)
    elif choice == '3':
        summary = get_summary(income_list, expense_list)
        print("\nSummary")
        print("Total Income:", summary['total_income'])
        print("Total Expenses:", summary['total_expenses'])
        print("Balance:", summary['balance'])
        print("Income by Category:", summary['income_by_category'])
        print("Expenses by Category:", summary['expenses_by_category'])
    elif choice == '4':
        break
    else:
        print("Invalid choice. Please try again.")
```
