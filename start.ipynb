```python
# Initialize an empty dictionary to store account balances
accounts = {}

def deposit(account_number, amount):
    if account_number in accounts:
        accounts[account_number] += amount
        print(f"Deposited {amount} into account {account_number}.")
    else:
        print(f"Account {account_number} does not exist.")

def withdraw(account_number, amount):
    if account_number in accounts:
        if accounts[account_number] >= amount:
            accounts[account_number] -= amount
            print(f"Withdrew {amount} from account {account_number}.")
        else:
            print(f"Insufficient balance in account {account_number}.")
    else:
        print(f"Account {account_number} does not exist.")

def view_account(account_number):
    if account_number in accounts:
        print(f"Account Number: {account_number}")
        print(f"Balance: {accounts[account_number]}")
    else:
        print(f"Account {account_number} does not exist.")

# Main program loop
while True:
    print("\nMenu:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. View Account")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        account_num = input("Enter account number: ")
        amount = float(input("Enter amount to deposit: "))
        deposit(account_num, amount)
    elif choice == "2":
        account_num = input("Enter account number: ")
        amount = float(input("Enter amount to withdraw: "))
        withdraw(account_num, amount)
    elif choice == "3":
        account_num = input("Enter account number: ")
        view_account(account_num)
    elif choice == "4":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number from 1 to 4.")
```
