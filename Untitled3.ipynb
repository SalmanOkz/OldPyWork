```python
import pygame
import random

# Define constants
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
GROUND_WIDTH = 600
GROUND_HEIGHT = 400
GOAL_WIDTH = 100
GOAL_HEIGHT = 50
PLAYER_SIZE = 20
BALL_SIZE = 10
PLAYER_SPEED = 5
BALL_SPEED = 7
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Define player class
class Player(pygame.sprite.Sprite):
    def _init_(self, x, y, team):
        super()._init_()
        self.image = pygame.Surface((PLAYER_SIZE, PLAYER_SIZE))
        self.image.fill(WHITE)
        self.rect = self.image.get_rect()
        self.rect.center = (x, y)
        self.team = team
        self.dx = 0
        self.dy = 0

    def update(self):
        self.rect.x += self.dx
        self.rect.y += self.dy
        self.check_boundaries()

    def check_boundaries(self):
        self.rect.left = max(self.rect.left, 0)
        self.rect.right = min(self.rect.right, SCREEN_WIDTH)
        self.rect.top = max(self.rect.top, 0)
        self.rect.bottom = min(self.rect.bottom, SCREEN_HEIGHT)

# Define ball class
class Ball(pygame.sprite.Sprite):
    def _init_(self):
        super()._init_()
        self.image = pygame.Surface((BALL_SIZE, BALL_SIZE))
        self.image.fill(WHITE)
        self.rect = self.image.get_rect()
        self.rect.center = (SCREEN_WIDTH // 2, SCREEN_HEIGHT // 2)
        self.dx = random.choice([-1, 1]) * BALL_SPEED
        self.dy = random.choice([-1, 1]) * BALL_SPEED

    def update(self):
        self.rect.x += self.dx
        self.rect.y += self.dy
        self.check_boundaries()

    def check_boundaries(self):
        if self.rect.left <= 0 or self.rect.right >= SCREEN_WIDTH:
            self.dx *= -1
        if self.rect.top <= 0 or self.rect.bottom >= SCREEN_HEIGHT:
            self.dy *= -1

# Initialize Pygame
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
clock = pygame.time.Clock()

# Create player and ball groups
all_sprites = pygame.sprite.Group()
players = pygame.sprite.Group()
ball_group = pygame.sprite.Group()

# Create players
player1 = Player(100, SCREEN_HEIGHT // 2, "red")
player2 = Player(700, SCREEN_HEIGHT // 2, "blue")
all_sprites.add(player1, player2)
players.add(player1, player2)

# Create ball
ball = Ball()
all_sprites.add(ball)
ball_group.add(ball)

# Game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Player movement
    keys = pygame.key.get_pressed()
    player1.dx, player1.dy = 0, 0
    if keys[pygame.K_UP]:
        player1.dy = -PLAYER_SPEED
    if keys[pygame.K_DOWN]:
        player1.dy = PLAYER_SPEED
    if keys[pygame.K_LEFT]:
        player1.dx = -PLAYER_SPEED
    if keys[pygame.K_RIGHT]:
        player1.dx = PLAYER_SPEED

    # Update sprites
    all_sprites.update()

    # Drawing
    screen.fill(BLACK)
    all_sprites.draw(screen)
    pygame.display.flip()

    clock.tick(60)

pygame.quit()
```
